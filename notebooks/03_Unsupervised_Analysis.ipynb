```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

errors = np.load("../artifacts/reconstruction_errors.npy")

sns.histplot(errors, bins=50, kde=True)
plt.axvline(errors.mean() + 3 * errors.std(), color='orange', linestyle='--', label='Threshold')
plt.title("Reconstruction Error Distribution")
plt.xlabel("Error")
plt.legend()
plt.show()
```
